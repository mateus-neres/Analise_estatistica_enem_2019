#**AREA DE AMOSTRAGEM SERÁ A CIDADE DE CAMPINAS-SP**

#Criando nova tabela com a amostragem de Campinas-SP

```python
enem_campinas = enem_sp.loc[enem_sp.MUNICIPIO_RESIDENCIA == "Campinas"]
```

```python
enem_campinas.head()
```

```python
enem_campinas.shape
```

##*AMOSTRAGEM ALEATÓRIA SIMPLES**

Seleção execultadas por meio de sorteio sem nenhum filtro

```python
np.random.seed(1) # Função seed serve para amarsenar o valor coletado aleatoriamente pelo python no valor desejado (1)
```

###Separando a amostra com 20% do total da tabela enem_campinas com a função .sample(frac=0.20)

```python
amostra_simples = enem_campinas.sample(frac=0.20) # Porcentagem
```

```python
amostra_simples.head(5)
```

```python
amostra_simples.shape
```

###Podemos também coletar uma quantidade determinada de registro apenas digitando o valor de restros que desejamos .sample(3000)

```python
amostra_simples = enem_campinas.sample(3000) # Quantidade numérica
```

```python
amostra_simples.head(5)
```

```python
amostra_simples.shape
```

###Outra forma de obter uma amostragem

```python
np.random.seed(2)
```

```python
amostra = np.random.choice(a=[0,1], size=13198, replace=True, p=[0.8,0.2])
```

#**AMOSTRAGEM SISTEMÁTICA**

```python
enem_campinas.shape
```

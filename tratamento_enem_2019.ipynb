#**Renomear varias colunas**

```python
# Aplicando em apenas uma coluna
dados1 = dados1.rename(columns={"NU_NOTA_REDACAO":"NOTA_REDACAO"})
```

```python
# Aplicando em varias colunas
dados1 = dados1.rename(columns={"NU_NOTA_CN":"NOTA_CN", "NU_NOTA_CH":"NOTA_CH",
                                "NU_NOTA_LC":"NOTA_LC", "NU_NOTA_MT":"NOTA_MT"})
```

```python
dados1 = dados1.rename(columns={"NU_INSCRICAO":"INSCRICAO","NO_MUNICIPIO_RESIDENCIA":"MUNICIPIO_RESIDENCIA",
                                "NU_IDADE":"IDADE", "TP_SEXO":"SEXO","TP_ESTADO_CIVIL":"ESTADO_CIVIL",
                                "TP_COR_RACA":"RACA", "TP_NACIONALIDADE":"NACIONALIDADE","TP_ESCOLA":"ESCOLA","Q025":"INTERNET"})
```

```python
dados1 = dados1.rename(columns={"NU_NOTA_COMP1":"COMP1","NU_NOTA_COMP2":"COMP2","NU_NOTA_COMP3":"COMP3","NU_NOTA_COMP4":"COMP4","NU_NOTA_COMP5	":"COMP5"})
```

```python
dados1 = dados1.rename(columns={"TP_LINGUA":"LINGUA"})
```

```python
dados1.head()
```

#**Conrrigindo erro nas notas individuais**

```python
dados1.loc[:,"NOTA_CN"] /= 10
dados1.loc[:,"NOTA_CH"] /= 10
dados1.loc[:,"NOTA_LC"] /= 10
dados1.loc[:,"NOTA_MT"] /= 10
```

```python
display(dados1.head(5))
```

##**Reanomear registros (linha)**

```python
dados1["RACA"] = dados1["RACA"].replace({0:"não_declarado", 1:"branca", 2:"preta",3:"parda",4:"amarela",5:"indígena"})
dados1["INTERNET"] = dados1["INTERNET"].replace({"A":"não", "B":"sim"})
# Outra maneira de fazer:
# Exp: dados1["INTERNET"] = dados1["INTERNET"].replace(["A","B"], ["não","sim"])
```

```python
dados1["ESCOLA"] = dados1["ESCOLA"].replace([1,2,3,4], ["não_respondeu","pública","privada","exterior"])
```

```python
dados1["LINGUA"] = dados1["LINGUA"].replace([0,1],["inglês","espanhol"])
```

```python
dados1.head()
```

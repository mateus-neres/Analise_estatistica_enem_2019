##Função de probabilidade de ocorrer um evento


```python
def probab (A,E):
  resultado = (A / E)*100
  print(f"{resultado:.2f}")
```

```python
probab(len(enem_guarulhos), len(enem_sp))
```

```python
probab(len(enem_bauru), len(enem_sp))
```

##Probabilidade de não ocorrer um evento

```python
def n_probab(A, B):
  resultado = (1 - (A / B)) * 100
  print(f"{resultado:.2f}")
```

```python
n_probab(len(enem_bauru), len(enem_sp))
```

```python
n_probab(len(enem_guarulhos), len(enem_sp))
```

##Probabilidade da união do multamente exclusivos(A U B)

```python
# CRIANDO A FUNÇÃO PROBABILIDADE DA UNIÃO MUTUAMENTE EXCLUSIVOS (AUB)
def probab_uniao (A,B,E):
  resultado = (A/E + B/E) * 100
  print(f"{resultado:.2f}")
```

```python
probab_uniao(len(enem_guarulhos), len(enem_bauru), len(enem_sp))
```

##Probabilidade da intersecção de eventos

```python
# cRIANDO A FUNÇÃO PROBABILIDADE DA INTERSECÇÃO DE EVENTOS
def probab_intersec (A, B, E):
  produto = (A/E * B/E)*100
  print(f"{produto:.3f}")
```

```python
# PROBABILIDADE DE RETIRAR UM VESTIBULANDO DE GUARULHOS E BAURO (COM REPOSIÇÃO)
probab_intersec(len(enem_guarulhos), len(enem_bauru), len(enem_sp))
```
